```python
import pandas as pd
pd.__version__
```

```python
students = {
    "name": ['John', 'Bob', 'Eve', 'Eric', 'Jane'],
    "quiz": [90, 92, 96, 91, 14],
    "mid": [91, 92, 94, 96, 12],
    "final": [88, 91, 93, 99, 13],
    "attendance": [9, 9, 8, 9, 5]
}

df  = pd.DataFrame(students)
df
```

```python
df['total'] = df['quiz'] + df['mid'] + df['final'] + df['attendance']
df['average'] = df[['quiz', 'mid', 'final']].mean(axis=1)
df['passed'] = df['average'] >= 50
df
```

```python
df[df['passed'] == True]
```

```python
df[df['attendance'] < 6]
```
