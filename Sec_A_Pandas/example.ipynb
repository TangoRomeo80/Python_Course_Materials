```python
import pandas as pd
pd.__version__
```

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'Ash', 'Asif'],
    'quiz': [85, 10, 78, 92, 88],
    'mid': [80, 15, 82, 88, 90],
    'final': [90, 12, 85, 95, 91],
    'attendance': [10, 9, 8, 10, 9]
}

df = pd.DataFrame(data)
df
```

```python
df.tail(2) # Print the last 2 rows of the dataframe
```

```python
df.columns # Get the column names of the dataframe
```

```python
df.dtypes # Get the data types of each column in the dataframe
```

```python
df.info() # Get a concise summary of the dataframe, including data types and non-null counts
```

```python
df.describe() # Get summary statistics for the numeric columns in the dataframe
```

```python
df['total'] = df['quiz'] + df['mid'] + df['final'] + df['attendance']
df['average'] = df[['quiz', 'mid', 'final', 'attendance']].mean(axis=1)
df['passed'] = df['average'] >= 50
df
```

```python
df[df['passed'] == True]
```

```python
df[df['attendance'] < 9]
```

```python
df[['name', 'average']]
```

```python
df.sort_values(by='average', ascending=False)
```
