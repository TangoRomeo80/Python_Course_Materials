```python
import pandas as pd
pd.__version__
```

```python
students = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'quiz': [85, 90, 78, 92, 18],
    'mid': [88, 85, 80, 95, 20],
    'final': [90, 87, 82, 96, 18],
    'attendance': [9, 9, 8, 8, 5]
}

df = pd.DataFrame(students)
df
```

```python
df['total'] = df['quiz'] + df['mid'] + df['final'] + df['attendance']
df
```

```python
df['average'] = df[['quiz', 'mid', 'final', 'attendance']].mean(axis=1)
df['passed'] = df['average'] >= 50
df
```

```python
df[df['passed'] == True]
```

```python
df[df['attendance'] < 6]
```
