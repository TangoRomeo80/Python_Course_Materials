```python
import pandas as pd
```

```python
tips = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
tips = pd.read_csv(tips)
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
tips.groupby('day')['total_bill'].mean()
```

```python
tips.groupby(['day', 'time'])[['total_bill', 'tip']].mean()
```

```python
import matplotlib.pyplot as plt

avg_tip_by_day = tips.groupby('day')['tip'].mean()

avg_tip_by_day.plot(kind='bar')
plt.title('Average Tip by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip')
plt.xticks(rotation=0)
plt.show()
```

```python
avg_bill_by_day = tips.groupby('day')['total_bill'].mean()
# coloring multiple bars
avg_bill_by_day.plot(kind='bar', color=['blue', 'orange', 'green', 'red'])
plt.title('Average Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill')
plt.xticks(rotation=0)
plt.show()
```
