### Dataframe is a 2-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object.

```python
# Example of dataframe.
data ={
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'quiz': [85, 10, 78, 92, 88],
    'mid': [80, 15, 75, 90, 82],
    'final': [90, 15, 80, 88, 91],
    'attendance': [10, 2, 8, 10, 9]
}

students_df = pd.DataFrame(data)
print(students_df)
```

```python
# Basic inspections
print(students_df.head())  # First 5 rows
print(students_df.head(3))  # First 3 rows
print(students_df.tail(2))  # Last 2 rows
```

```python
print(students_df.shape)  # Number of rows and columns
print(students_df.columns)  # Column names
print(students_df.dtypes)  # Data types of each column
```

```python
print(students_df.info())  # Summary of the DataFrame
print(students_df.describe())  # Statistical summary of numerical columns
```

```python
# Creating new columns
students_df["total"] = students_df["quiz"] + students_df["mid"] + students_df["final"]
students_df["average"] = students_df[['quiz', 'mid', 'final']].mean(axis=1) # axis=1 means we want to calculate the mean across columns for each row.
students_df['passed'] = students_df['average'] >= 50
students_df['passed_status'] = students_df['passed'].map({True: 'Passed', False: 'Failed'})
print(students_df)
```

```python
# Filtering data
students_df[students_df['passed'] == True]
students_df[students_df['attendance'] < 5]
# Only selecting specific columns
students_df[['Name', 'average', 'passed_status']]
```

```python
students_df.sort_values(by='average', ascending=False)
```

```python
students_df.to_csv('students_results.xlsx', index=False)
```
