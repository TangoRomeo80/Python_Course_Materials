```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
tips = pd.read_csv(tips_url)
tips.head()
```

```python
avg_tip_by_day = tips.groupby("day")["tip"].mean()
avg_tip_by_day.plot(kind="bar")
# Customize colors
plt.bar(avg_tip_by_day.index, avg_tip_by_day.values, color=["blue", "orange", "green", "red"])  
plt.title("Average Tip by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Average Tip")
plt.show()
```

```python
avg_bill_by_day = tips.groupby("day")["total_bill"].mean()
avg_bill_by_day.plot(kind="bar")
plt.bar(avg_bill_by_day.index, avg_bill_by_day.values, color=["cyan", "magenta", "yellow", "gray"])
plt.title("Average Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill")
plt.show()
```

```python
tips.plot(x="total_bill", y="tip", kind="scatter")
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
smoker_counts = tips["smoker"].value_counts()
smoker_counts.plot(kind="pie", autopct="%1.1f%%", colors=["lightblue", "lightcoral"])
plt.title("Smoker vs Non-Smoker Distribution")
plt.ylabel("")  # Hide y-label for better aesthetics
plt.show()
```
